```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the 3 cleaned data
benin_data=pd.read_csv("../data/benin_clean.csv")
sieraleone_data=pd.read_csv("../data/sierraleone_data.csv")
togo_data=pd.read_csv("../data/togo_data_clean.csv")
```

```python
#Adding Column Country for each country data
sieraleone_data['country']='Sierra Leone'
benin_data['country'] = 'Benin'
togo_data['country']='Togo'
```

```python
# merging country datasets
merged_data = pd.concat([sieraleone_data,benin_data,togo_data])
```

```python
#Checking the merged Dataset
merged_data.tail()
```

#### Task 2. Metric Comparison
-- Boxplots of GHI, DNI, DHI side-by-side (one plot per metric, colored by country).
Summary Table comparing mean, median, and standard deviation of GHI, DNI, DHI across countries.

```python
# metrics Column 'GHI','DNI','DHI']

metrics=['GHI','DNI','DHI']
for metric in metrics:
    plt.figure (figsize=(8,6))
    sns.boxplot(data=merged_data,x='country',y=metric, palette='Set2')
    plt.title(f'{metric} DIstribution by country')
    plt.show()
```

##### depending on the graph the Benin country has high compared to other 2 country, Togo and sierra leone, the DNI and GHI

### makign Summary Table based on thos 3 metrics, ('GHI','DNI','DHI' )

```python
summary_table = merged_data.groupby('country')[metrics].agg(['mean','median','std']).round(2)
summary_table
```

So from the summary table it indicates that the benin data has high value of those metrics compared to the rest
### here is some report based on the table 


- Benin shows the highest mean GHI and DNI, indicating strong direct and global solar potential.

- Sierra Leone has the lowest GHI and DNI, suggesting lower solar generation potential overall.

- Togo has a balanced profile, with GHI and DNI values close to Benin and the highest DHI median, indicating steady diffuse irradiance.
